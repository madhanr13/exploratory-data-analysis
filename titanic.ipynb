##### Import libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### load the dataset

```python
df = pd.read_csv('titanic.csv')
df
```

##### analyzing the dataset

```python
# df.isnull().sum()
df.info()
```

```python
# checking unique values
df.nunique()
```

```python
# check if there's any duplicated values
df.duplicated().sum()
```

##### removing missing values

```python
df.drop(["Cabin"], axis=1, inplace=True)
```

```python
df['Age'].fillna(df["Age"].mean(), inplace=True)
```

```python
df
```

##### replacing values

```python
df["Sex"] = df["Sex"].replace("male", "Male")
df["Sex"] = df["Sex"].replace("female", "Female")
df
```

##### changing the datatypes and adding new column

```python
df["Age"] = df["Age"].astype(int)
df
```

```python
# adding a new column

df["Family_Count"] = df["SibSp"] + df["Parch"]
df
```

#### Data visualization

```python
# bar plot

plt.figure(figsize=(3, 2))
sns.barplot(x="Sex", y="Survived", data=df)
plt.title("Gender vs Survived")
plt.xlabel("Gender")
plt.ylabel("Survived")
plt.show()
```

```python
# histogram

plt.figure(figsize=(3, 2))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```
